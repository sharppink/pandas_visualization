```python
data4 = pd.read_excel('capstone/주민등록인구(지도, 년, 월)_.xlsx')
data4
```

```python
data4 = data4.melt(id_vars='행정구역별', var_name='날짜', value_name='인구수')
data4['날짜'] = pd.to_datetime(data4['날짜'], format='%Y.%m')
data4 = data4.rename(columns={'행정구역별': '지역'})
```

```python
data4
```

```python
mapping = {
    '서울특별시': '서울',
    '부산광역시': '부산',
    '대구광역시': '대구',
    '인천광역시': '인천',
    '광주광역시': '광주',
    '대전광역시': '대전',
    '울산광역시': '울산',
    '세종특별자치시': '세종',
    '경기도': '경기',
    '강원특별자치도': '강원',
    '충청북도': '충북',
    '충청남도': '충남',
    '전북특별자치도': '전북',
    '전라남도': '전남',
    '경상북도': '경북',
    '경상남도': '경남',
    '제주특별자치도': '제주',
}
data4['지역'] = data4['지역'].replace(mapping)
data4
```

```python
final_df = pd.merge(merged_df, data4, on=['날짜', '지역'], how='left')
final_df = final_df.drop(columns=['거래량_y'])
final_df = final_df.rename(columns={'거래량_x': '거래량'})
final_df
```

```python
final_df.to_excel('통합.xlsx', index=False)
```

```python
df = pd.read_excel('capstone/소비자물가지수.xlsx')
```

```python
df['년도'] = df['년도'].astype(str).str.replace('o', '0', regex=False)
```

```python
df['날짜'] = pd.to_datetime(df['년도'], format='%Y.%m', errors='coerce')
```

```python
df_long = df.melt(id_vars='날짜', var_name='지역', value_name='물가지수')
```

```python
# 1. '날짜' 컬럼은 id_vars로 유지, '년도' 컬럼은 제외하고 melt 수행
value_vars = [col for col in df.columns if col not in ['날짜', '년도']]
df_long = df.melt(id_vars='날짜', value_vars=value_vars, var_name='지역', value_name='물가지수')

# 2. '지역'에 '년도'가 들어간 잘못된 행 제거
df_long = df_long[df_long['지역'] != '년도']

# 3. '물가지수'를 숫자형으로 변환 (문자열로 인식된 경우 대비)
df_long['물가지수'] = pd.to_numeric(df_long['물가지수'], errors='coerce')

# 4. 인덱스 재정렬
df_long = df_long.reset_index(drop=True)
```

```python
df_long
```

```python
final_data = pd.merge(final_df, df_long, on=['날짜', '지역'], how='left') 
final_data
```

```python
final_data.to_excel('통합.xlsx', index=False)
```
