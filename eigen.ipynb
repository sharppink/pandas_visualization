```python
import numpy as np
```

```python
A = np.array([[ 6, -2, -1],
              [-2, 6, -1],
              [-1, -1, 5]])
```

```python
val, vet = np.linalg.eig(A)
```

```python
print('val :')
print(val)
```

```python
print('vet :')
print(vet)
```

```python
B = np.linalg.inv(A)
B
```
