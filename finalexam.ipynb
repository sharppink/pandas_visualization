```python
import numpy as np
```

```python
fish_data = np.array([2,3,3,4,4,4,4,5,5,6])
fish_data
```

```python
len(fish_data)
```

```python
N = len(fish_data)
sum_value = np.sum(fish_data)
mu = sum_value/ N
mu
```

```python
var = np.sum((fish_data - mu) **2)/N
var
```

```python
np.var(fish_data)
np.var(fish_data, ddof=0) #ddof = 1 표본분산, 0 모분산
```
