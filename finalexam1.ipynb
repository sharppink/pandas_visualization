```python
import numpy as np
import pandas as pd
import scipy as sp
from scipy import stats
```

```python
np.random.seed(1)
sampling_norm = stats.norm.rvs(loc=100, scale=5, size = 1500)
len(sampling_norm)
```

```python
np.sum(sampling_norm >= 100)
```

```python
def sum(samping_norm) :
    for i in range(0, len(sampling_norm)):
        if (sampling_norm[i] >= 100):
            sum = sum + sampling_norm
            i += i

    return sum
```
