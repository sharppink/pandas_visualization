```python
data1_3 = pd.read_csv('/content/yemoonsaBigdata/datasets/Part5/ex7/07.01.03-apartment_prices_dataset.csv')
data1_3
```

```python
data1_3.isnull().sum()
```

```python
data1_3.describe()
```

```python
data1_3.isnull().sum()
```

```python
cols =data1_3.columns
cols
```

```python
var_count = {}

for col in cols:
  q1, q3 = data1_3[col].quantile([0.25, 0.75])
  iqr = q3 - q1
  count = (((data1_3[col]) < (q1 - 1.5*iqr)) | ((data1_3[col]) > (q3 + 1.5*iqr))).sum()
  var_count[col] = count

print(var_count)
```

```python
sorted_var_count = sorted(
    var_count.items(),
    key=lambda x: x[1],
    reverse=True
)

print(sorted_var_count)
```

```python
data3 = pd.read_csv('/content/yemoonsaBigdata/datasets/Part5/ex7/07.01.03-apartment_prices_dataset.csv')
data3
```

```python
data3 = data3.drop('Price', axis=1)
cols = data3.columns
print(cols)
```

```python
var_counts = {}

for col in cols:
  q1 = data3[col].quantile(0.25)
  q3 = data3[col].quantile(0.75)
  iqr = q3-q1
  var_counts[col] = (((data3[col]) < (q1 - 1.5*iqr))|((data3[col]) > (q3 + 1.5*iqr))).sum()

print(var_counts)
```

```python
sorted(var_counts.items(), key=lambda x: x[1], reverse=True)
```
