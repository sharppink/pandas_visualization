```python
data = data.replace('\n',' ')
data = data.replace('       ',' ')
data
```

```python
numbers = list(map(float, data.split()))
len(numbers)
```

```python
numbers = list(map(float, data.split()))
reshaped = np.array(numbers).reshape(-1,4)
df = pd.DataFrame(reshaped, columns=['inputs', 'method', 'time', 'knowledge'])
df
```

```python
def run_anova(df, dependent_var, factors):
    formula = f"{dependent_var} ~ " + " + ".join([f"C({factor})" for factor in factors])
    model = ols(formula, data=df).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    return anova_table
```

```python
anova_time = run_anova(df, "time", ["inputs", "method"])
anova_knowledge = run_anova(df, "knowledge", ["inputs", "method"])
```

```python
print("=== Time 변수 ANOVA ===")
print(anova_time)

print("\n=== Knowledge 변수 ANOVA ===")
print(anova_knowledge)
```

```python
model_time = ols('time ~ C(inputs) + C(method)', data=df).fit()
anova_time = sm.stats.anova_lm(model_time, typ=2)

print("\n=== time 변수 ANOVA ===")
print(anova_time)

model = ols('knowledge ~ C(inputs)', data=df).fit()
anova_result = sm.stats.anova_lm(model, typ=2)
print("\n=== inputs 변수 ANOVA ===")
print(anova_result)
```

```python
model = ols('knowledge ~ C(inputs) * C(method)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print(anova_table)
```

```python
maov = MANOVA.from_formula('time + knowledge ~ C(inputs) + C(method)', data=df)
print(maov.mv_test())
```

```python
independent_vars = ['inputs', 'method', 'time']

for var in independent_vars:
    print(f"\n=== {var}에 대한 ANOVA 결과 ===")
    if var in ['inputs', 'method']:
        model = ols(f'knowledge ~ C({var})', data=df).fit()
    else:
        model = ols(f'knowledge ~ {var}', data=df).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    print(anova_table)
```

```python
independent_vars = ['inputs', 'method', 'time']

# 각 변수에 대해 ANOVA 수행 후 p-value만 출력
for var in independent_vars:
    if var in ['inputs', 'method']:
        model = ols(f'knowledge ~ C({var})', data=df).fit()
    else:
        model = ols(f'knowledge ~ {var}', data=df).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    p_value = anova_table["PR(>F)"][0]
    print(f"{var}에 대한 모형 p-value: {p_value:.4f}")
```

```python
from scipy.stats import t
```

```python
model = ols('knowledge ~ C(method)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# 사후검정: LSD 계산
means = df.groupby("method")["knowledge"].mean()
n = df.groupby("method").size().min()  # 각 그룹의 최소 샘플 수
MSE = anova_table.loc["Residual", "sum_sq"] / anova_table.loc["Residual", "df"]
df_error = anova_table.loc["Residual", "df"]
t_critical = t.ppf(1 - 0.05 / 2, df_error)  # 양측검정 (alpha=0.05)

# LSD 계산
LSD = t_critical * np.sqrt(2 * MSE / n)

print(f"\n[그룹별 평균값]")
print(means)
print(f"\n[LSD 값]: {LSD:.4f}")
```
