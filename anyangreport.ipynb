```python
reshaped = np.array(numbers).reshape(-1,4)
df = pd.DataFrame(reshaped, columns=['inputs', 'method', 'time', 'knowledge'])
df
```

```python
model = ols('knowledge ~ C(method)', data=df).fit()
anova_result = sm.stats.anova_lm(model, typ=2)
```
