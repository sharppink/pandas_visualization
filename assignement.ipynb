```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
from scipy import stats
from statsmodels.formula.api import ols
```

```python
a = np.array([ 2, 3, 4, 3, 6, 4, 2, 3, 4, 3, 4, 3, 4, 2, 6, 4, 3, 4, 5, 6, 4, 3])
b = np.array([ 4, 5, 3, 3, 4, 5, 6, 4, 5, 4, 3, 4, 2, 5, 6, 4, 5, 6, 7, 5, 4, 6, 5, 6 ])
c = np.array([6,5,7,8,6,5,4,6,5,6,5,6,7,6,5,5,6])
```

```python
f_statistic, p_value = stats.f_oneway(a, b, c)

# 결과 출력
anova_table = pd.DataFrame({
    'Source': ['Between Groups', 'Within Groups'],
    'DF': [3 - 1, len(a) + len(b) + len(c) - 3],
    'F': [f_statistic, ''],
    'p-value': [p_value, '']
})

print(anova_table)
```

```python
data = [a, b, c]
all_data = np.concatenate(data)
group_labels = ['a'] * len(a) + ['b'] * len(b) + ['c'] * len(c)
group_means = [np.mean(g) for g in data]
grand_mean = np.mean(all_data)

# 제곱합 계산
SS_between = sum(len(g) * (np.mean(g) - grand_mean)**2 for g in data)
SS_within = sum(((x - np.mean(g))**2).sum() for x, g in zip(data, data))
SS_total = SS_between + SS_within

# 자유도
df_between = len(data) - 1
df_within = len(all_data) - len(data)
df_total = df_between + df_within

# 평균제곱
MS_between = SS_between / df_between
MS_within = SS_within / df_within

# F 통계량
F = MS_between / MS_within

# p-value
p_value = stats.f.sf(F, df_between, df_within)

# 결과 출력
anova_table = pd.DataFrame({
    'Source': ['Between Groups', 'Within Groups', 'Total'],
    'SS': [SS_between, SS_within, SS_total],
    'df': [df_between, df_within, df_total],
    'MS': [MS_between, MS_within, ''],
    'F': [F, '', ''],
    'p-value': [p_value, '', '']
})

# precision 설정 (수정된 부분)
pd.set_option('display.precision', 4)
print(anova_table)
```
