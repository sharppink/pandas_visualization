```python
data['v4이용종류'] = 0
data['v4_1이용여부'] = 0
data
```

```python
def v4(row):
    if row['v41이용종류'] ==1.0 :
        return 1
    if row['v42이용종류'] ==1.0 :
        return 2
    if row['v43이용종류'] ==1.0 :
        return 3
```

```python
def v4_1(row):
    if row['v41이용종류'] ==1.0 :
        return 1
    if row['v42이용종류'] ==1.0 :
        return 1
    if row['v43이용종류'] ==1.0 :
        return 0
```

```python
data['v4이용종류'] = data.apply(v4, axis=1)
data['v4_1이용여부'] = data.apply(v4_1, axis =1)
data
```

```python
from scipy.stats import ttest_ind
```

```python
columns_to_test = data.columns[data.columns.get_loc("v5이용목적"):data.columns.get_loc("v34안전 수칙 위반 처벌 및 단속 강화 필요 여부") + 1]
```

```python
data['v4_1이용여부'].unique()
```

```python
use = data[data['v4_1이용여부'] ==1]
n_use = data[data['v4_1이용여부'] ==0]
results = []
```

```python
for col in columns_to_test:
    m_values = use[col].dropna()
    f_values = n_use[col].dropna()

    if len(m_values) > 1 and len(f_values) > 1:
        t_stat, p_value = ttest_ind(m_values, f_values, equal_var=False)
        results.append({
            '항목': col,
            '이용자 평균': round(m_values.mean(), 3),
            '비이용자 평균': round(f_values.mean(), 3),
            'p값': round(p_value, 5),
            '유의한 차이': p_value < 0.05
        })
```

```python
ttest_df = pd.DataFrame(results)
print(ttest_df)
```

```python
from scipy.stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
use_levels = sorted(data['v4이용종류'].dropna().unique())
```

```python
significant_columns = []
```

```python
for col in columns_to_test:
    groups = [data[data['v4이용종류'] == use][col].dropna() for use in use_levels]
    if all(len(g) > 1 for g in groups):
        f_stat, p_value = f_oneway(*groups)
        if p_value < 0.05:
            print(f" [ANOVA 유의] {col} | F={round(f_stat, 4)}, p={round(p_value, 5)}")
            significant_columns.append(col)
for col in significant_columns:
    print(f"\n [Tukey HSD 결과] {col}")
    tukey = pairwise_tukeyhsd(endog=data[col].dropna(), groups=data['v4이용종류'][data[col].notna()], alpha=0.05)
    print(tukey.summary())
```

```python
from scipy.stats import chi2_contingency
import numpy as np
```

```python
chi2_results = []
for col in columns_to_test:
    ct = pd.crosstab(data['v4_1이용여부'], data[col])
    if ct.shape[0] > 1 and ct.shape[1] > 1:
        chi2, p, dof, expected = chi2_contingency(ct)
        chi2_results.append({
            '변수': col,
            '카이제곱 통계량': round(chi2, 4),
            'p값': round(p, 5),
            '자유도': dof,
            '유의함': p < 0.05
        })

# 데이터프레임 변환 및 정렬
chi2_df = pd.DataFrame(chi2_results).sort_values(by='p값').reset_index(drop=True)
chi2_df
```
