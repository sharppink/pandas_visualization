```python
import pandas as pd
```

```python
data = pd.read_excel('project.xlsx')
data
```

```python
data['v4이용종류'] = 0
data
```

```python
def v4(row):
    if row['v41이용종류'] ==1.0 :
        return 1
    if row['v42이용종류'] ==1.0 :
        return 2
    if row['v43이용종류'] ==1.0 :
        return 3
```

```python
data['v4이용종류'] = data.apply(v4, axis=1)
```

```python
data.drop(columns=['v41이용종류','v42이용종류','v43이용종류'], inplace=True)
data
```

```python
from scipy.stats import ttest_ind
```

```python
columns_to_test = data.columns[data.columns.get_loc("v5이용목적"):data.columns.get_loc("v34안전 수칙 위반 처벌 및 단속 강화 필요 여부") + 1]
```

```python
male = data[data['v1성별'] ==1]
female = data[data['v1성별'] ==2]
results = []
```

```python
for col in columns_to_test:
    m_values = male[col].dropna()
    f_values = female[col].dropna()

    if len(m_values) > 1 and len(f_values) > 1:
        t_stat, p_value = ttest_ind(m_values, f_values, equal_var=False)
        results.append({
            '항목': col,
            '남성 평균': round(m_values.mean(), 3),
            '여성 평균': round(f_values.mean(), 3),
            'p값': round(p_value, 5),
            '유의한 차이': p_value < 0.05
        })
```

```python
ttest_df = pd.DataFrame(results)
print(ttest_df)
```

```python
from scipy.stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
grade_levels = sorted(data['v2학년'].dropna().unique())
```

```python
significant_columns = []
```

```python
for col in columns_to_test:
    groups = [data[data['v2학년'] == grade][col].dropna() for grade in grade_levels]
    if all(len(g) > 1 for g in groups):
        f_stat, p_value = f_oneway(*groups)
        if p_value < 0.05:
            print(f" [ANOVA 유의] {col} | F={round(f_stat, 4)}, p={round(p_value, 5)}")
            significant_columns.append(col)
```

```python
for col in significant_columns:
    print(f"\n [Tukey HSD 결과] {col}")
    tukey = pairwise_tukeyhsd(endog=data[col].dropna(), groups=data['v2학년'][data[col].notna()], alpha=0.05)
    print(tukey.summary())
```
