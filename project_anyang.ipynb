```python
import pandas as pd
```

```python
data = pd.read_excel('project.xlsx')
data
```

```python
data['v4이용종류'] = 0
data
```

```python
def v4(row):
    if row['v41이용종류'] ==1.0 :
        return 1
    if row['v42이용종류'] ==1.0 :
        return 2
    if row['v43이용종류'] ==1.0 :
        return 3
```

```python
data['v4이용종류'] = data.apply(v4, axis=1)
```

```python
data.drop(columns=['v41이용종류','v42이용종류','v43이용종류'], inplace=True)
data
```
