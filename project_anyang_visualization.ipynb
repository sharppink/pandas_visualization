```python
import pandas as pd
```

```python
data = pd.read_excel('project.xlsx')
data
```

```python
gender_count = data['v1성별'].value_counts().sort_index()
gender_count
```

```python
gender_count_name = gender_count.rename(index = {1:'남성',2:'여성'}).reset_index()
gender_count_name.columns = ['성별', '응답자 수']
gender_count_name.to_csv("성별_응답자_분포.csv", index=False, encoding='utf-8-sig')
```

```python
univ_count = data['v3소속대학'].value_counts().sort_index()
univ_count = univ_count.rename(index = {1:'사회과학대학', 2:'신학대학', 3:'예술대학', 4:'인문대학', 5:'창의융합대학'})
univ_count.columns = ['소속대학', '응답자 수']
univ_count.to_csv("소속대학.csv", index=True, encoding='utf-8-sig')
```

```python
grad_count = data['v2학년'].value_counts().sort_index()
grad_count
```

```python
grad_count = grad_count.rename(index= { 1:'1학년', 2:'2학년', 3:'3학년', 4:'4학년'})
grad_count.columns = ['학년', '응답자 수']
grad_count.to_csv('학년.csv', index = True, encoding='utf-8')
```
