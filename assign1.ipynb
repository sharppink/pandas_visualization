```python
import numpy as np
from scipy.stats import f, chi2
```

```python
# 표본 크기
n1 = 20
n2 = 30
p = 2  # 변수 개수: 국어, 영어

# 평균 벡터
x1_bar = np.array([78, 83])
x2_bar = np.array([84, 80])

# 공분산 행렬
S1 = np.array([[33, 10],
               [10, 28]])

S2 = np.array([[32, 23],
               [23, 42]])
```

```python
# 공통 공분산 행렬
Sp = ((n1 - 1) * S1 + (n2 - 1) * S2) / (n1 + n2 - 2)

# 평균 차이
diff = x1_bar - x2_bar

# T² 계산
T2_equal = (n1 * n2) / (n1 + n2) * diff @ np.linalg.inv(Sp) @ diff

# F-분포로 변환
F_stat = ((n1 + n2 - p - 1) * T2_equal) / ((n1 + n2 - 2) * p)
df1 = p
df2 = n1 + n2 - p - 1
p_value_equal = 1 - f.cdf(F_stat, df1, df2)

print("1) 공분산이 같다고 가정:")
print(f"T² = {T2_equal:.4f}")
print(f"F({df1}, {df2}) = {F_stat:.4f}")
print(f"p-value = {p_value_equal:.4f}")

```

```python
# 각 공분산을 표본 크기로 나눠 더함
S_diff = S1 / n1 + S2 / n2

# T² 계산
T2_unequal = diff @ np.linalg.inv(S_diff) @ diff

# p-value using chi-square distribution
p_value_unequal = 1 - chi2.cdf(T2_unequal, df=p)

print("\n2) 공분산이 다르다고 가정:")
print(f"T² = {T2_unequal:.4f}")
print(f"Chi-square({p}) = {T2_unequal:.4f}")
print(f"p-value = {p_value_unequal:.4f}")
```
