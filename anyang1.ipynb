```python
xsum =0; sqsum=0
x=[10,12,21]
for i in range(0,3):
    xsum = xsum + x[i]
    sqsum=sqsum+x[i]**2

mean = xsum/len(x)
var = sqsum/(len(x)-1)-(len(x)*mean**2)/(len(x)-1)
sd = math.sqrt(var)
print('mean:',round(mean,3), '분산:',round(var,3), 'sd:',round(sd,3))
```

```python
def reg(x,y):
    xsum = 0; ysum =0; xsqusm=0;  ysqusm=0; xysum=0
    a=range(0, len(x))
    n=len(x)
    for i in a:
        xsum = xsum + x[i]
        xsqusm = xsqusm + x[i]**2
        ysum = ysum + y[i]
        ysqusm = ysqusm + y[i]**2
        xysum = xysum + x[i] *y[i]
    xmean=xsum/n
    ymean=ysum/n

    b1 = (xysum-n*xmean*ymean)/(xsqusm-n*xmean**2)
    b0 = (ymean-b1*xmean)

    if(b1>0):
        print('추정된 회귀식=',round(b0,3), '+',round(b1,3), 'x')
    elif(b1<0):
        print('추정된 회귀식=',round(b0,3), '-', round(b1,3), 'x')
    else:
        print('x')
```

```python
n = 10
p = 0.3
binom = scipy.stats.binom(n, p)

print(round(binom.pmf(3), 3))
print(round(1 - binom.cdf(5), 3))
print(round(binom.mean(), 3))
print(round(binom.var(), 3))
print(binom.rvs(10))
```
