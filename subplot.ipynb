```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
fig = make_subplots(rows =1, cols =2)

fig.add_trace(
    go.Scatter(x=[1,2,3], y=[4,5,6]),
    row=1, col=1
)
fig.add_trace(
    go.Scatter(x=[20,30,40], y = [50,60,70]),
               row = 1 , col =2
)
```

```python
fig = make_subplots(rows =1, cols =1)
fig.append_trace(go.Scatter(
    x=[3,4,5],
    y=[1000,1100,1200],
), row=1, col=1)
fig.append_trace(go.Scatter(
    x=[2,3,4],
    y=[100,110,120],
), row=1, col=1)
fig.append_trace(go.Scatter(
    x=[0,1,2],
    y=[10,11,12],
), row=1, col=1)
fig.show()
```

```python
fig = make_subplots(rows =3, cols =1)
fig.append_trace(go.Scatter(
    x=[3,4,5],
    y=[1000,1100,1200],
), row=1, col=1)
fig.append_trace(go.Scatter(
    x=[2,3,4],
    y=[100,110,120],
), row=2, col=1)
fig.append_trace(go.Scatter(
    x=[0,1,2],
    y=[10,11,12],
), row=3, col=1)
fig.show()
```

```python
fig = make_subplots(rows=2, cols=2, start_cell='bottom-left')
fig.add_trace(go.Scatter(x=[1,2,3], y=[4,5,6]),
              row=1,col=1)
fig.add_trace(go.Scatter(x=[20,30,3], y=[10,20,30]),
              row=1,col=2)
fig.add_trace(go.Scatter(x=[100,200,300], y=[40,50,60]),
              row=2,col=1)
fig.add_trace(go.Scatter(x=[10,20,30], y=[400,500,600]),
              row=2,col=2)
fig.show()
```

```python
fig = make_subplots(rows =1 , cols=2)

fig.add_trace(
    go.Scatter(
        x=[1,2,3],
        y=[4,5,6],
        mode = 'markers+text',
        text = ['a','b','c'],
        textposition='bottom center'
    ),
    row=1, col =1
)
fig.add_trace(
    go.Scatter(
        x=[20,30,40],
        y=[50,60,70],
        mode = 'markers+text',
        text = ['d','e','f'],
        textposition='bottom center'
    ),
    row=1, col =2
)
fig.show()
```

```python
labels = ['1','2','3','4','5']
specs = [[{'type':'domain'}, {'type': 'domain'}], [{'type':'domain'},{'type':'domain'}]]
fig = make_subplots(rows = 2, cols =2, specs=specs)

fig.add_trace(go.Pie(labels = labels, values=[25,25,25,25], 
                     title='first pie'), row =1 ,col =1)
fig.add_trace(go.Pie(labels = labels, values=[10,20,30,40], 
                     title='second pie'), row =1 ,col =2)
fig.add_trace(go.Pie(labels = labels, values=[50,50], 
                     title='third pie'), row =2 ,col =1)
fig.add_trace(go.Pie(labels = labels, values=[20,20,20,20,20], 
                     title='fourth pie'), row =2 ,col =2)
fig.show()
```
